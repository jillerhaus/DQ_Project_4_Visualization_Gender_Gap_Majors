```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

stem_cats = ['Psychology', 'Biology','Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']

fig = plt.figure(figsize = (16, 20))
for i_stem in range(0,6):
    ax = fig.add_subplot(6,3,i_stem*3+1)
    ax.plot(women_degrees['Year'],women_degrees[stem_cats[i_stem]], c = cb_dark_blue, label = 'Women', lw = 3)
    ax.plot(women_degrees['Year'],100-women_degrees[stem_cats[i_stem]], c = cb_orange, label = 'Men', lw = 3)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(stem_cats[i_stem])
    for location in ['top','bottom','left','right']:
        ax.spines[location].set_visible(False)
    ax.tick_params(top = 'off', bottom = 'off', left = 'off', right = 'off', labelbottom = 'off')
    if i_stem == 0:
        ax.text(2002, 82, 'Women')
        ax.text(2005, 15, 'Men')
    elif i_stem == 5:
        ax.text(2005, 88, 'Men')
        ax.text(2002, 8, 'Women')
        ax.tick_params(labelbottom = 'on')

#2nd row
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
for i_lib in range(0,5):
    ax = fig.add_subplot(6,3,(i_lib*3)+2)
    ax.plot(women_degrees['Year'],women_degrees[lib_arts_cats[i_lib]], c = cb_dark_blue, label = 'Women', lw = 3)
    ax.plot(women_degrees['Year'],100-women_degrees[lib_arts_cats[i_lib]], c = cb_orange, label = 'Men', lw = 3)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(lib_arts_cats[i_lib])
    for location in ['top','bottom','left','right']:
        ax.spines[location].set_visible(False)
    ax.tick_params(top = 'off', bottom = 'off', left = 'off', right = 'off', labelbottom = 'off')
    if i_lib == 0:
        ax.text(2002, 74, 'Women')
        ax.text(2005, 22, 'Men')
    elif i_lib == 4:
        ax.tick_params(labelbottom = 'on')
        
#3rd row
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture', 'Business', 'Architecture']
for i_other in range(0,6):
    ax = fig.add_subplot(6,3,(i_other*3)+3)
    ax.plot(women_degrees['Year'],women_degrees[other_cats[i_other]], c = cb_dark_blue, label = 'Women', lw = 3)
    ax.plot(women_degrees['Year'],100-women_degrees[other_cats[i_other]], c = cb_orange, label = 'Men', lw = 3)
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_title(other_cats[i_other])
    for location in ['top','bottom','left','right']:
        ax.spines[location].set_visible(False)
    ax.tick_params(top = 'off', bottom = 'off', left = 'off', right = 'off', labelbottom = 'off')
    if i_other == 0:
        ax.text(2002, 90, 'Women')
        ax.text(2005, 5, 'Men')
    elif i_other == 5:
        ax.text(2005, 64, 'Men')
        ax.text(2002, 32, 'Women')
        ax.tick_params(labelbottom = 'on')
   
plt.show()
plt.savefig('gender_degrees.png')
    
```
